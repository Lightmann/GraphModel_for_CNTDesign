```python
#data_org = pd.read_excel('data_all.xlsx') 
data_org = pd.read_excel('data_all_back_A.xlsx') # 2020.06.06 use data back -- no log(*)
graph_array = data_load('graph_array_368.dat')
```

```python
ids = [0, # C->Y
       8,64,
       21,14,22,
       90,88,154,152,
       367 #C->S, C->R, C->A, CARS->T, ST->Y
      ]
```

```python
graphs = [graph_array[i] for i in ids]
```

```python
# gm_array_array = []
gm_array = []
for _ in range(20):
#     data_back = random_resample(data_org,shuffle=True)
#     gm_array = []
    for G in graphs:
        print('fit ',_, ': ',nx2str(G))
#         G = str2nx(s)
        data_back = random_resample(data_org,shuffle=True)
        gm = GraphModel_cnt(G=G,idy='Y',renamer=cnt_renamer,
                            assign_submodel=assign_submodel_cnt,
                            gb_estimators=GB_ESTIMATORS,
                            n_iter_search=GB_SEARCH_ITER,
                            submodel_cv=CV,
                           )
        data = data_back.dropna(subset=gm.renamer(G.nodes),how='any').reset_index(drop=True)
        gm.fit(data,n_estimator=20,ratio=0.9,reuse=False)
        gm_array.append(gm)
    
#     gm_array_array.append(gm_array)
```
